```python
from scipy.special import gamma
import numpy as np
import matplotlib.pyplot as plt
```

```python
def h(n, x, t):
    return 1 / (n + 1)**0.25 * (2 * t)**(0.5 * (n + x)) / np.sqrt(gamma(n + x))
```

```python
x = 3
t = 3
n_grid = np.arange(20)

fig, ax = plt.subplots()
ax.plot(n_grid, h(n=n_grid, x=x, t=t))
```

```python
x_grid = np.arange(20)
t = 2
n = 20

fig, ax = plt.subplots()
ax.plot(n_grid, h(n=20, x=x_grid, t=t))
```

```python
x = 4
t = 2
n_grid = np.arange(40)

fig, ax = plt.subplots()
ax.plot(n_grid, h(n=n_grid, x=x, t=t) / h(n=n_grid + 1, x=x + 1, t=t))
```
