```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import perm
from math import factorial
```

```python
def double_factorial(k):
    return np.arange(1, k + 1, 2).prod()

def chi2_cf(u, T):
    return 1 / np.sqrt(1 - 2 * u * T)
```

```python
T = 1
u = np.linspace(-0.4, 0, 1000) * 1j

k_arr = np.arange(100)
cf_sig = np.zeros((k_arr.size, u.size), dtype=complex)
for k in range(k_arr.size):
    if not k:
        cf_sig[k] = 1
    else:
        cf_sig[k] = cf_sig[k - 1] + (u * T / 2)**k * perm(2 * k, 2 * k - k) / factorial(k)
```

```python
fig, ax = plt.subplots()
ax.plot(np.imag(u), chi2_cf(u, T), 'g')
for k in k_arr[20:]:
    ax.plot(np.imag(u), cf_sig[k])
ax.grid()
```
